```python
df_player_game = df_init.iloc[:,:25]
df_player_game.head()
```

```python
main_variables = df_player_game.groupby('Player', as_index=False).agg({'Year': 'count', 'Age': 'max','Tm':'count','FantPt':'sum','G':'sum','GS':'sum','Cmp':'sum','Att':'sum','Yds':'sum','TD':'sum','Int':'sum','Att.1':'sum','Yds.1':'sum','Y/A':'sum','TD.1':'sum','Rec':'sum','Yds.2':'sum','Y/R':'sum','TD.2':'sum','Height (inches)':'mean','Weight':'mean'})
main_variables.sort_values(by='FantPt')
main_variables
```

```python
main_variables.sort_values(by='FantPt',ascending=False).head(20)

```

```python
main_variables['Avg FP Per Game'] = main_variables['FantPt'] / main_variables['G']
```

```python
main_variables.sort_values(by='Avg FP Per Game',ascending=False).head(20)
```
